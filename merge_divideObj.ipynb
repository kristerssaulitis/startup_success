```python
objects = pd.read_csv("./raw_data/objects.csv")
relationships = pd.read_csv("./raw_data/relationships.csv")
people = pd.read_csv("./raw_data/people.csv")
funding_rounds = pd.read_csv("./raw_data/funding_rounds.csv")
ipos = pd.read_csv("./raw_data/ipos.csv")
investments = pd.read_csv("./raw_data/investments.csv")
funds = pd.read_csv("./raw_data/funds.csv")
degrees = pd.read_csv("./raw_data/degrees.csv")
acquisitions = pd.read_csv("./raw_data/acquisitions.csv")
```

```python
objects_collumns = objects.drop(['entity_id','normalized_name', 'permalink', 'homepage_url', 'logo_url', 'logo_width', 'logo_height', 'state_code', 'created_by', 'first_milestone_at', 'last_milestone_at', 'milestones'], axis=1)
relationships_collumns = relationships.drop(['id', 'relationship_id', 'sequence','created_at', 'updated_at'], axis=1)
people_collumns = people.drop(['id', 'birthplace'], axis=1)
funding_rounds_collumns = funding_rounds.drop(['id','pre_money_valuation', 'pre_money_currency_code', 'post_money_valuation', 'post_money_currency_code', 'source_url', 'source_description', 'created_by', 'created_at', 'updated_at'], axis=1)
ipos_collumns = ipos.drop(['id','ipo_id', 'source_url', 'source_description', 'created_at', 'updated_at'], axis=1)
investments_collumns = investments.drop(['id', 'created_at', 'updated_at'], axis=1)
funds_collumns = funds.drop(['id', 'fund_id', 'source_url', 'source_description', 'created_at', 'updated_at'], axis=1)
degrees_collumns = degrees.drop(['id', 'created_at', 'updated_at'], axis=1)
acquisitions_collumns = acquisitions.drop(['id', 'source_url', 'source_description', 'created_at', 'updated_at'], axis=1)
# offices_collumns = offices.drop(['id','description', 'office_id', 'address1', 'address2', 'zip_code', 'state_code', 'latitude', 'longitude', 'created_at', 'updated_at'], axis=1)

```

```python
people_degrees = pd.merge(people_collumns, degrees_collumns, left_on = 'object_id', right_on = 'object_id', how = 'left')
investments_funds = pd.merge(investments_collumns, funds_collumns, left_on = 'investor_object_id', right_on = 'object_id', how = 'left')
investments_funds_collumn = investments_funds.drop(['object_id'], axis=1)
objects_offices = pd.merge(objects_collumns, offices_collumns, left_on = 'id', right_on = 'object_id', how = 'left')
relationships_people = pd.merge(relationships_collumns, people_degrees, left_on = 'person_object_id', right_on = 'object_id', how = 'left')
relationships_people_collumn = relationships_people.drop(['object_id'], axis=1)
```

```python
objects_collumns.to_csv('./raw_data/new_obj.csv', index=False)
relationships_collumns.to_csv('./raw_data/new_relationships.csv', index=False)
people_collumns.to_csv('./raw_data/new_people.csv', index=False)
funding_rounds_collumns.to_csv('./raw_data/new_funding_rounds.csv', index=False)
ipos_collumns.to_csv('./raw_data/new_ipos.csv', index=False)
investments_collumns.to_csv('./raw_data/new_investments.csv', index=False)
funds_collumns.to_csv('./raw_data/new_funds.csv', index=False)
degrees_collumns.to_csv('./raw_data/new_degrees.csv', index=False)
acquisitions_collumns.to_csv('./raw_data/new_acquisitions.csv', index=False)
objects_offices.to_csv('./raw_data/new_objects_offices.csv', index=False)
```
